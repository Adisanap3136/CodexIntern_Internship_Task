```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the CSV file
file_path = 'Downloads/sales.csv'  # Replace with your file's actual path
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
data.head()
```

```python
# Calculate the average of the AMOUNTINFO column
average_amount = data['AMOUNTINFO'].mean()
print(f"The average order amount is: {average_amount:.2f}")
```

```python
# Convert ORDERDATE to datetime for better analysis
data['ORDERDATE'] = pd.to_datetime(data['ORDERDATE'])
```

```python
# Aggregate data for total sales amount per day
daily_sales = data.groupby('ORDERDATE')['AMOUNTINFO'].sum()

# Plot the bar chart
plt.figure(figsize=(12, 6))
daily_sales.plot(kind='bar', color='skyblue')
plt.title('Total Sales Amount Per Day', fontsize=14)
plt.xlabel('Order Date', fontsize=12)
plt.ylabel('Total Sales Amount', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Aggregate data for total sales amount per day
daily_sales = data.groupby('ORDERDATE')['AMOUNTINFO'].sum()

# Display the aggregated data as a table
display(daily_sales.reset_index())

# Plot the bar chart
plt.figure(figsize=(12, 6))
daily_sales.plot(kind='bar', color='skyblue')
plt.title('Total Sales Amount Per Day', fontsize=14)
plt.xlabel('Order Date', fontsize=12)
plt.ylabel('Total Sales Amount', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Display a sample of the relevant columns
display(data[['USERID', 'AMOUNTINFO']].sample(10))

# Scatter Plot: USERID vs. AMOUNTINFO
plt.figure(figsize=(12, 6))
plt.scatter(data['USERID'], data['AMOUNTINFO'], alpha=0.5, color='purple')
plt.title('Scatter Plot of USERID vs. AMOUNTINFO', fontsize=14)
plt.xlabel('USERID', fontsize=12)
plt.ylabel('Order Amount (AMOUNTINFO)', fontsize=12)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Generate a correlation matrix
correlation_matrix = data[['USERID', 'AMOUNTINFO']].corr()

# Display the correlation matrix as a table
display(correlation_matrix)

# Heatmap: Correlation Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Heatmap of Correlation Matrix', fontsize=14)
plt.tight_layout()
plt.show()
```
